```python
from tkinter import *
from tkinter import ttk
import pandas as pd
df = pd.read_excel(r"mycar.xlsx")
df = df.set_index(["Car","Model","Year"])
win =Tk()
win .title("DC RENTAL VALUER") 
photo = PhotoImage(file = "Dynamo_Logo.png")
win.iconphoto(False, photo)
win.geometry = ("500,700")
win.resizable(False,False)
win.config(bg="#14AACF")

mytoyota =["Sienna","Camry","Corolla","Venza"]
myhonda = ["Accord","Civic","CRV"]
mylexus = ["ES350","GX460"]
mychevrolet= ["Impala","Corvette","Silverado","Surburban","Tahoe"]
mykia = ["Sportage","Carnival","Stinger","Sorento","Forte"]
myyear= list(range(2017,2024))
 
def insurance():
    pass

def estimate():
    try:
        brand = brands.get()
        model = models.get()
        year =int(years.get()) # Converted to Integer because Year is Int DataType in source file
        hour = float(hours.get())
        price = df.loc[(brand,model,year),"Price"]
        flag = 0
        price1 = round(price * hour,2)
        if  insurance_value.get()==1:
            newprice = round(price1 *1.10,2)
            price_plus_tax =  round(newprice + (newprice*0.15),2)
        elif insurance_value.get()==2:
            newprice = round(price1 * 1.25,2)
            price_plus_tax =  round(newprice + (newprice*0.15),2)
        elif insurance_value.get()==3:
            newprice = round(price1 ,2)
            price_plus_tax =  round(newprice + (newprice*0.15),2)
            flag =1

        if flag == 0:
            answer= "Initial Price : $"+ str(price1)+ "\nPrice with insurance : $" + str(newprice) + "\nFinal price after tax : $" + str(price_plus_tax)
            answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")

        else:
            answer ="Initial Price : $" + str(price1) + "\nFinal price after tax : $"+ str(price_plus_tax)
            answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
    except:
        answer = "Invalid entry entered or you \nhaven't selected an insurance option"
        answerlabel.config(text=answer,font=("Arial,25,bold"),bg="white",fg="black")
        
# Function to clear the old values of the widget's parameters
def clear():
    var1.set("")
    var2.set("")
    var3.set("")
    hours.delete(0, END) 
    answerlabel.config(text="")
    insurance_value.set("")

# Bind the brand to model and year dropdown boxes 
def select_model(e):
    if brands.get()=="Toyota":
        models.config(value = mytoyota)
        years.config(value=myyear)
    elif brands.get()== "Honda":
         models.config(value = myhonda) 
         years.config(value=myyear)
    elif brands.get()== "Lexus":
         models.config(value = mylexus)
         years.config(value=myyear)
    elif brands.get()== "Chevrolet":
         models.config(value = mychevrolet)
         years.config(value=myyear)
    elif brands.get()== "Kia":
         models.config(value = mykia)
         years.config(value=myyear)
    
# Creation of the App's Frames     
frame1 =Frame(win,width=25,height=100,bg="white")
frame1.pack(side=TOP)
frame2 =Frame(win,width=500,height=100,bg="#14AACF")
frame2.pack(side=TOP)
frame3 =Frame(win,width=500,height=100,bg="#14AACF")
frame3.pack(side=TOP)
frame4 =Frame(win,width=500,height=100,bg="#14AACF")
frame4.pack(side=TOP)
frame5 =Frame(win,width=500,height=100,bg="#14AACF")
frame5.pack(side=TOP)
frame6 =Frame(win,width=500,height=100,bg="#14AACF")
frame6.pack(side=TOP)
frame7 =Frame(win,width=500,height=150,bg="#14AACF")
frame7.pack(side=TOP)

# Creation of the App's Widgets
answerlabel = Label(frame1,bg="white",fg="black",width=35,height=3,bd=2,relief=RAISED)
answerlabel.grid(row=0,column=0)

labelbrand=Label(frame2,text ="Brand",width=11,height=1,bd=2,relief=RAISED,bg="#14AACF")
labelbrand.config(font=("Arial,6"))
labelbrand.grid(row=0,column=0,pady=10,padx=5)

var1= StringVar()
brands = ttk.Combobox(frame2, width = 11, textvariable = var1,font=("Arial,6"))
brands.config(height=500)
brands.grid( row = 0,column = 1,pady=10,padx=5)
brands.current()
brands.bind("<<ComboboxSelected>>",select_model)

brands['values'] = ("Toyota", "Honda", "Lexus","Chevrolet","Kia") # Adding combobox drop down list
brands.config(state="readonly") # Prevent Manual Entry of values

label_model=Label(frame3,text ="Model",width=11,height=1,bg="#14AACF",bd=2,relief=RAISED)
label_model.grid(row=0,column=0,pady=10,padx=5)
label_model.config(font=("Arial,6"))

var2= StringVar()
models= ttk.Combobox(frame3, width = 11,height=5, textvariable = var2,font=("Arial,6"))
models.grid( row = 0,column = 1,pady=10,padx=5)
models.config(state="readonly") # Prevent Manual Entry of values

label_year=Label(frame4,text ="Year",width=11,height=1,bg="#14AACF",bd=2,relief=RAISED)
label_year.grid(row=0,column=0,pady=10,padx=5)
label_year.config(font=("Arial,6"))
var3= StringVar()
years= ttk.Combobox(frame4, width = 11,height=5, textvariable = var3,font=("Arial,6"))
years.grid( row = 0,column = 1,pady=10,padx=5)
years.config(state="readonly") #Prevent Manual Entry of values

hourlabel=Label(frame5,text ="No of Hours",width=11,height=1,bg="#14AACF",bd=2,relief=RAISED)
hourlabel.config(font=("Arial,6"))
hourlabel.grid(row=0,column=0,pady=10,padx=7)
hours = Entry(frame5,font=("Arial,6"),width=12,bg="white") 
hours.grid(row=0,column=1,pady=10,padx=7)

ins_label=Label(frame6,text ="Insurance Options",width=25,height=1,bg="#14AACF",
                font =("Arial,5,bold"),bd=2,relief=RAISED)
ins_label.grid(row=0,column=0,columnspan=3,padx=15)

insurance_value = IntVar()  
basic = Radiobutton(frame6, text="Basic", variable=insurance_value , value=1,  
                  bg="#14AACF",command=insurance)  
basic.grid(row=1,column=0)

comprehensive = Radiobutton(frame6, text="Comprehensive", variable=insurance_value , value=2,  
                  bg="#14AACF",command=insurance)  
comprehensive.grid(row=1,column=1,padx=5)
  
not_needed = Radiobutton(frame6, text="Not Needed", variable=insurance_value , value=3,  
                  bg="#14AACF",command=insurance)  
not_needed.grid(row=1,column=2,padx=5)

button_estimate=Button(frame7, text="SHOW PRICE",width=11,command=estimate,
                    bg="#AABDCC",fg= "green",font =("Arial,6,bold"),relief=RAISED,bd=2)
button_estimate.config(activebackground="#AABDCC",activeforeground="green")
button_estimate.grid(row=0,column=0,padx=5,pady=10)

button_clear=Button(frame7, text="CLEAR",width=11,command=clear,
                      bg="#AABDCC",fg= "red",font =("Arial,6,bold"),relief=RAISED,bd=2)
button_clear.config(activebackground="#AABDCC",activeforeground="red")
button_clear.grid(row=0,column=1,padx=5,pady=10)

win.mainloop()
```
